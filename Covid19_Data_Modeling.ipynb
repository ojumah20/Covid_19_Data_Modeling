```python
#droping the columns we do not need 
dimstate = dimstate.drop(['Id','Id2','date','state','cases','deaths'],axis = 1)
```

```python
#droping duplicates 
dimstate = dimstate.drop_duplicates()
```

```python
dimstate.head()
```

```python
#join dimstate dataframe to state_daily dataframe to get the state short code
dimstate=pd.merge(dimstate, state_daily, left_on='fips', right_on='fips', how='left')
```

```python
dimstate.head()
```

```python
#to select only the columns we need
dimstate = dimstate[['County','State','Population Estimate 2018','fips','state']]
```

```python
#to drop duplicates
dimstate = dimstate.drop_duplicates()
dimstate.head()
```

```python
dimstate.rename(columns = {'Population Estimate 2018':'County_Population_Estimate','fips':'State_Id','state': 'State_short_code'},inplace=True)
```
