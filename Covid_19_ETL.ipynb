```python
#select date column to be used for creating dimenstion table for date
data3["lastModified"]
```

```python
data3.head()
```

```python
#view the newly created date column
data3["pk_date"]
```

```python
#create a new dataframe from the new date column
new_date = pd.DataFrame(data3['pk_date'], columns=["pk_date"])
new_date.head()
```

```python
#covert the column pk_date from string to datetime 
new_date['pk_date'] = pd.to_datetime(new_date["pk_date"])
new_date['pk_date'].head()
```

```python
#extract day from the date column
new_date['day'] = new_date['pk_date'].dt.day
new_date
```

```python
#extract month from the date column
new_date['month'] = new_date['pk_date'].dt.month
new_date
```

```python
#extract year from the date column
new_date['year'] = new_date['pk_date'].dt.year
new_date
```

```python
#drop duplicate dates
new_date.drop_duplicates()
```
